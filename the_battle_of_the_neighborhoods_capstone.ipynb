## 01. Introduction
---
### Identifying the Business Problem
Toronto is home to the Maple Leafs, one of seven NHL hockey teams in Canada. The Maple Leafs were founded in 1917 and die-hard NHL fans know that they have enjoyed great success as well as their fair share of hardships. The Leafs are known as one of the “Original Six” and have a large fan following.

Being a historical franchise, as well as having their home ice in the largest city in Canada, makes attending a game at Scotiabank Arena a highly sought after experience. People will travel from all over Canada, the US, and the world to do so.

Scotiabank Arena is located in Downtown Toronto making it a hot spot for tourism. When planning your next trip to attend a Maple Leafs game, the options can be overwhelming. Whether it is booking a hotel, deciding on a restaurant, or simply grabbing a coffee to go, the last thing you want to experience is analysis paralysis. It is easy to experience decision fatigue when planning your trip to downtown Toronto.


### Target Audience
The goal of this project will be to eliminate the need for transportation and is focused on those who want to be within walking distance of both the arena and dining. Whether it is a quick cup of coffee to start your day or choosing from the most highly recommended restaurants close to Scotiabank Arena, I aim to streamline the decision making process.

My results will be narrowed to the top seven recommended choices within each of the categories that are explored. The number seven was not chosen at random. Psychologist George Miller’s study proved seven to be the magic number for how many options our mind can process at a single time. Therefore, when viewing the results you will find seven of the top recommended hotels, coffee shops, and restaurants within a reasonable walking distance of 1.5 miles from Scotiabank Arena.
